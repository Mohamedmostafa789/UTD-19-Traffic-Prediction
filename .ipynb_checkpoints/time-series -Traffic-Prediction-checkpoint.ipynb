#### **2.2 SARIMA Model**
SARIMA handles seasonality, trends, and noise in time series data.


```python

def preprocess_time_series(df, target_column, exog_columns, train_end, test_start, freq='H'):
    """
    Preprocesses time series data by downsampling, handling missing values, and splitting into train/test sets.
    
    Parameters:
        df (pd.DataFrame): The input DataFrame containing the time series data.
        target_column (str): The target variable column name.
        exog_columns (list): List of column names to use as exogenous variables.
        train_end (str): End datetime for the training data.
        test_start (str): Start datetime for the testing data.
        freq (str): Frequency for downsampling (default is 'H' for hourly).
    
    Returns:
        train_data, test_data, train_exog, test_exog (pd.Series, pd.Series, pd.DataFrame, pd.DataFrame)
    """
    # Ensure the index is a DatetimeIndex
    if not isinstance(df.index, pd.DatetimeIndex):
        if 'datetime' in df.columns:
            df.set_index('datetime', inplace=True)
        else:
            raise ValueError("DataFrame must have a DatetimeIndex or a 'datetime' column.")

    # Downsample target variable to specified frequency
    target_downsampled = df[target_column].resample(freq).mean()

    # Resample exogenous variables to match the same frequency
    exog_downsampled = df[exog_columns].resample(freq).mean()

    # Handle missing and infinite values
    exog_downsampled = exog_downsampled.replace([np.inf, -np.inf], np.nan)
    exog_downsampled.fillna(method='ffill', inplace=True)
    exog_downsampled.fillna(method='bfill', inplace=True)

    # Ensure indexes match
    exog_downsampled = exog_downsampled.loc[target_downsampled.index]

    # Split into training and testing sets
    train_data = target_downsampled[:train_end]
    test_data = target_downsampled[test_start:]
    train_exog = exog_downsampled[:train_end]
    test_exog = exog_downsampled[test_start:]

    # Final alignment
    train_exog = train_exog.loc[train_data.index]
    test_exog = test_exog.loc[test_data.index]

    return train_data, test_data, train_exog, test_exog
```

```python
def train_and_evaluate_sarimax(train_data, test_data, train_exog, test_exog, order, seasonal_order):
    """
    Trains a SARIMAX model and evaluates its performance.

    Parameters:
        train_data (pd.Series): Training data for the target variable.
        test_data (pd.Series): Testing data for the target variable.
        train_exog (pd.DataFrame): Training data for exogenous variables.
        test_exog (pd.DataFrame): Testing data for exogenous variables.
        order (tuple): SARIMAX (p, d, q) order.
        seasonal_order (tuple): Seasonal SARIMAX (P, D, Q, S) order.
    
    Returns:
        dict: A dictionary containing the model, predictions, MAE, and RMSE.
    """
    # Train the SARIMAX model
    model = SARIMAX(
        train_data,
        exog=train_exog,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    fit = model.fit(disp=False)

    # Forecast
    predictions = fit.forecast(steps=len(test_data), exog=test_exog)

    # Evaluate model performance
    mae = mean_absolute_error(test_data, predictions)
    rmse = mean_squared_error(test_data, predictions, squared=False)

    # Print results
    print(f"SARIMAX Model - MAE: {mae:.2f}, RMSE: {rmse:.2f}")

    return {
        "model": fit,
        "predictions": predictions,
        "mae": mae,
        "rmse": rmse
    }
```

```python

# Example usage
# Assuming your DataFrame is named `df` and includes 'datetime' as a column
target_column = 'traffic_flow'
exog_columns = ['holiday', 'incident', 'severity']  # Replace with your actual exogenous variables
train_end = '2016-10-28 23:55:00'
test_start = '2016-10-29 00:05:00'

# Preprocess data
train_data, test_data, train_exog, test_exog = preprocess_time_series(
    df,
    target_column=target_column,
    exog_columns=exog_columns,
    train_end=train_end,
    test_start=test_start
)

# Train and evaluate SARIMAX
results = train_and_evaluate_sarimax(
    train_data, test_data, train_exog, test_exog,
    order=(1, 1, 1), seasonal_order=(1, 1, 1, 24)
)

# Access predictions
predictions = results['predictions']
```

```python

# Function: Fine-tune SARIMA
def fine_tune_sarima(train_data, train_exog, param_ranges, seasonal_period, maxiter=500):
    p, d, q = param_ranges['p'], param_ranges['d'], param_ranges['q']
    P, D, Q = param_ranges['P'], param_ranges['D'], param_ranges['Q']
    s = [seasonal_period]

    parameter_combinations = list(itertools.product(p, d, q, P, D, Q, s))

    best_aic = float("inf")
    best_params = None
    best_model = None

    for params in parameter_combinations:
        try:
            model = SARIMAX(
                train_data,
                exog=train_exog,
                order=(params[0], params[1], params[2]),
                seasonal_order=(params[3], params[4], params[5], params[6]),
                enforce_stationarity=False,
                enforce_invertibility=False
            )
            result = model.fit(disp=False, maxiter=maxiter, method='powell')
            if result.aic < best_aic:
                best_aic = result.aic
                best_params = params
                best_model = result
        except Exception as e:
            print(f"Skipping parameters {params} due to error: {e}")
            continue

    if best_params is None:
        raise ValueError("No valid SARIMA parameters found during grid search.")

    print(f"Best SARIMA Parameters: {best_params}, AIC: {best_aic}")
    return best_params, best_model

def evaluate_sarima(fitted_model, test_data, test_exog):
    """
    Evaluates the SARIMA model on test data.

    Parameters:
        fitted_model: Fitted SARIMAX model.
        test_data (pd.Series): Testing target data.
        test_exog (pd.DataFrame): Testing exogenous variables.

    Returns:
        dict: Evaluation metrics and predictions.
    """
    try:
        forecast = fitted_model.forecast(steps=len(test_data), exog=test_exog)
    except Exception as e:
        raise ValueError(f"Error during SARIMA forecasting: {e}")

    mae = mean_absolute_error(test_data, forecast)
    rmse = mean_squared_error(test_data, forecast, squared=False)

    print(f"SARIMA Model - MAE: {mae:.2f}, RMSE: {rmse:.2f}")
    return {"forecast": forecast, "mae": mae, "rmse": rmse}

# Function: Train and Evaluate Prophet
def train_and_evaluate_prophet(df, train_end, test_start, target_column):
    """
    Trains and evaluates a Prophet model.

    Parameters:
        df (pd.DataFrame): DataFrame with a datetime index and target column.
        train_end (str): End datetime for training data.
        test_start (str): Start datetime for testing data.
        target_column (str): Target column name.

    Returns:
        dict: Prophet results including predictions, MAE, and RMSE.
    """
    # Prepare data for Prophet
    df = df.reset_index()
    df.rename(columns={"datetime": "ds", target_column: "y"}, inplace=True)

    # Split data
    train_data = df[df['ds'] <= train_end]
    test_data = df[df['ds'] >= test_start]

    # Initialize Prophet
    prophet_model = Prophet()
    prophet_model.fit(train_data)

    # Ensure future dates exactly match test data range
    future_dates = pd.DataFrame(test_data['ds'])
    future_dates.rename(columns={"ds": "ds"}, inplace=True)

    # Forecast
    forecast = prophet_model.predict(future_dates)

    # Align indices
    forecast.set_index('ds', inplace=True)
    test_data.set_index('ds', inplace=True)

    # Ensure alignment of predictions and test data
    common_index = test_data.index.intersection(forecast.index)
    aligned_test_data = test_data.loc[common_index]
    aligned_predictions = forecast.loc[common_index, 'yhat']

    # Debugging: Check alignment
    print(f"Test Data Length: {len(test_data)}")
    print(f"Forecast Length: {len(forecast)}")
    print(f"Common Index Length: {len(common_index)}")
    print(f"Aligned Test Data Length: {len(aligned_test_data)}")
    print(f"Aligned Predictions Length: {len(aligned_predictions)}")

    # Ensure lengths match
    if len(aligned_test_data) != len(aligned_predictions):
        raise ValueError("Aligned test data and predictions lengths do not match.")

    # Evaluate
    mae = mean_absolute_error(aligned_test_data['y'], aligned_predictions)
    rmse = mean_squared_error(aligned_test_data['y'], aligned_predictions, squared=False)

    print(f"Prophet Model - MAE: {mae:.2f}, RMSE: {rmse:.2f}")

    return {"forecast": aligned_predictions, "mae": mae, "rmse": rmse, "prophet_model": prophet_model}
```

# Prepare Data for Deep Learning
First, normalize and reshape the data into a 3D array format: (samples, time_steps, features).

```python
def prepare_data_for_deep_learning(df, target_column, exog_columns, lookback=24):
    """
    Prepare time series data for deep learning models.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing time series data.
        target_column (str): Target variable.
        exog_columns (list): Exogenous variable columns.
        lookback (int): Number of time steps to look back.
    
    Returns:
        generator (TimeseriesGenerator): Time series generator for training.
        scaler_target (MinMaxScaler): Scaler used for the target variable.
        scaler_exog (MinMaxScaler): Scaler used for exogenous variables.
    """
    # Ensure datetime index
    if not isinstance(df.index, pd.DatetimeIndex):
        raise ValueError("The DataFrame must have a DatetimeIndex.")

    # Normalize data
    scaler_target = MinMaxScaler()
    scaler_exog = MinMaxScaler()
    
    # Fit scalers
    target_scaled = scaler_target.fit_transform(df[[target_column]])
    exog_scaled = scaler_exog.fit_transform(df[exog_columns])
    
    # Combine scaled features
    combined_scaled = np.hstack([target_scaled, exog_scaled])
    
    # Create TimeseriesGenerator
    generator = TimeseriesGenerator(combined_scaled, target_scaled, length=lookback, batch_size=32)
    
    return generator, scaler_target, scaler_exog
```

###  Build Deep Learning Models
 LSTM, GRU, and TCN models with attention mechanisms.

LSTM Model with Attention

```python

def build_lstm_model(input_shape):
    """
    Build an LSTM model with an attention mechanism using the functional API.
    
    Parameters:
        input_shape (tuple): Shape of the input data (time_steps, features).
    
    Returns:
        model (tf.keras.Model): Compiled LSTM model.
    """
    # Define inputs
    inputs = Input(shape=input_shape)
    
    # LSTM layer
    lstm_output = LSTM(64, return_sequences=True)(inputs)
    
    # Attention mechanism
    query = lstm_output  # Query is the LSTM output
    value = lstm_output  # Value is also the LSTM output
    attention_output = Attention()([query, value])
    
    # Dense layers
    dense_output = Dense(32, activation='relu')(attention_output)
    outputs = Dense(1)(dense_output)  # Final output
    
    # Define the model
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    
    return model
```

# Step 2: Model Predictions vs Actual Values
Visualize how well the model's predictions align with the actual values.

```python
def plot_predictions(test_data, predictions, model_name):
    """
    Plot actual vs predicted values for a model.

    Parameters:
        test_data (pd.Series): Actual target values from the test set.
        predictions (np.ndarray): Model predictions.
        model_name (str): Name of the model for labeling the plot.
    """
    # Reshape predictions to 1D if necessary
    if predictions.ndim > 1:
        predictions = predictions[:, -1, 0]  # Extract the last time step and flatten

    # Ensure the lengths match
    if len(predictions) != len(test_data):
        predictions = predictions[-len(test_data):]
    plt.figure(figsize=(15, 7))
    plt.plot(test_data.index, test_data, label='Actual', alpha=0.7)
    plt.plot(test_data.index, predictions, label='Predicted', alpha=0.7)
    plt.title(f'{model_name} - Predictions vs Actual')
    plt.xlabel('Time')
    plt.ylabel('Traffic Flow')
    plt.legend()
    plt.grid()
    plt.show()

# Example: Plot for LSTM Predictions
plot_predictions(test_data, lstm_eval['predictions'], 'LSTM')
```

# Step 3: Error Analysis
Plot the errors to observe how they vary over time

```python
def plot_errors(test_data, predictions, model_name):
    """
    Plot residuals (errors) for a model.

    Parameters:
        test_data (pd.Series): Actual target values from the test set.
        predictions (np.ndarray): Model predictions.
        model_name (str): Name of the model for labeling the plot.
    """
    # Reshape predictions if needed
    if predictions.ndim > 1:
        predictions = predictions[:, -1, 0]  # Extract the last time step and flatten

    # Trim predictions to match the length of test_data
    predictions = predictions[:len(test_data)]

    # Calculate residuals
    residuals = test_data.values - predictions

    # Plot residuals
    plt.figure(figsize=(15, 7))
    plt.plot(test_data.index, residuals, label='Residuals', color='red')
    plt.title(f'{model_name} - Residuals Over Time')
    plt.xlabel('Time')
    plt.ylabel('Error')
    plt.legend()
    plt.grid()
    plt.show()
# Example: Plot for LSTM Predictions
plot_errors(test_data, lstm_eval['predictions'], 'LSTM')
```

##### GRU Model with Dropout

```python
from tensorflow.keras.layers import GRU, Dropout

def build_gru_model(input_shape):
    """
    Build a GRU model with dropout.
    
    Parameters:
        input_shape (tuple): Shape of the input data (time_steps, features).
    
    Returns:
        model (tf.keras.Model): Compiled GRU model.
    """
    model = Sequential([
        GRU(64, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        GRU(32),
        Dense(1)  # Final output
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

###### Temporal Convolutional Network (TCN)


```python
from tensorflow.keras.layers import Conv1D, Flatten

def build_tcn_model(input_shape):
    """
    Build a Temporal Convolutional Network (TCN).
    
    Parameters:
        input_shape (tuple): Shape of the input data (time_steps, features).
    
    Returns:
        model (tf.keras.Model): Compiled TCN model.
    """
    model = Sequential([
        Conv1D(64, kernel_size=3, activation='relu', input_shape=input_shape),
        Conv1D(32, kernel_size=3, activation='relu'),
        Flatten(),
        Dense(1)  # Final output
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

####  Train and Evaluate Models
Train each model and evaluate its performance

```python
def train_and_evaluate_model(model, train_generator, test_generator, epochs=50):
    """
    Train and evaluate a deep learning model.
    
    Parameters:
        model (tf.keras.Model): The model to train.
        train_generator (TimeseriesGenerator): Training data generator.
        test_generator (TimeseriesGenerator): Testing data generator.
        epochs (int): Number of epochs to train the model.
    
    Returns:
        history: Training history.
        evaluation: Model evaluation metrics.
    """
    # Train model
    history = model.fit(train_generator, epochs=epochs, validation_data=test_generator, verbose=1)

    # Evaluate model
    evaluation = model.evaluate(test_generator)
    print(f"Test Loss: {evaluation[0]:.4f}, Test MAE: {evaluation[1]:.4f}")
    
    return history, evaluation
```
